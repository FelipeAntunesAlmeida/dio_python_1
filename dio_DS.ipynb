```python
#eliminando colunas desnecessárias para a análise que será feita e armazenando em uma nova variável
df_limpo=df.drop(columns=["DT_GERACAO","HH_GERACAO","ANO_ELEICAO","CD_MUNICIPIO","CD_MUN_SIT_BIOMETRICA","CD_GENERO","CD_ESTADO_CIVIL","CD_FAIXA_ETARIA","CD_GRAU_ESCOLARIDADE"])
```

```python
df_limpo.head() #exibindo o dataframe após eliminar colunas
```

```python
df_limpo.head(10)
```

```python
#verificando a quantidade de eleitores e o grau de escolaridade ordenados pela quantidade

escolaridade=df_limpo.groupby("DS_GRAU_ESCOLARIDADE")["QT_ELEITORES_PERFIL"].sum()
escolaridade=escolaridade.sort_values(ascending=False)
escolaridade
```

```python
pd.options.display.float_format = '{:8,.0f}'.format
escolaridade.describe()
```

```python
escolaridade.plot.barh();
```

```python
escolaridade.describe()
```

```python
#quantidade por sexo

sexo=df_limpo.groupby("DS_GENERO")["QT_ELEITORES_PERFIL"].sum()
sexo=sexo.sort_values(ascending=False)
sexo
```

```python
#a mesma verificação por faixa etária

fx_etaria=df_limpo.groupby("DS_FAIXA_ETARIA")["QT_ELEITORES_PERFIL"].sum()
fx_etaria.sort_values(ascending=False)

```

```python
#50% da amostra está entre 21 e 79 anos
#as faixas etárias que estão acima da média está entre 21 e 69 anos
fx_etaria.describe()
```

```python
fx_etaria.plot.bar()
```
